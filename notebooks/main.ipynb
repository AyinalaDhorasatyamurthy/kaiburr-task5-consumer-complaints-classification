```python
import pandas as pd

# read the file with NO headers (so every row is treated as data)
df = pd.read_csv(
    "C:/Users/ASUS/Documents/consumer-complaints-classification/data/complaints.csv",
    low_memory=False,
    header=None,
    on_bad_lines='skip'
)

# show first few rows
df.head(3)
```

```python
df.columns = [
    "Date received",
    "Product",
    "Sub-product",
    "Issue",
    "Sub-issue",
    "Consumer complaint narrative",
    "Company public response",
    "Company",
    "State",
    "ZIP code",
    "Tags",
    "Consumer consent provided?",
    "Submitted via",
    "Date sent to company",
    "Company response to consumer",
    "Timely response?",
    "Consumer disputed?",
    "Complaint ID"
]

df.head()
```

```python
df.shape
```

```python
df = df[['Date received', 'Product', 'Issue', 'Consumer complaint narrative']]

# drop rows with missing text
df = df.dropna(subset=['Consumer complaint narrative']).reset_index(drop=True)

# rename columns for convenience
df = df.rename(columns={
    'Date received': 'date_received',
    'Product': 'product',
    'Issue': 'issue',
    'Consumer complaint narrative': 'text'
})

df.head()
```

```python
# map your products into the 4 target classes
TARGETS = {
    'Credit reporting, credit repair services, or other personal consumer reports': 'Credit reporting, repair, or other',
    'Debt collection': 'Debt collection',
    'Consumer loan': 'Consumer Loan',
    'Mortgage': 'Mortgage'
}

# keep only rows whose 'product' matches one of the keys
df = df[df['product'].isin(TARGETS.keys())].copy()

# map them to simplified names
df['label'] = df['product'].map(TARGETS)

# confirm counts per class
print("Class distribution:\n", df['label'].value_counts(), "\n")
df.head()
```

```python
print(df.isna().sum())       # check for missing values
print(df['label'].unique())  # confirm only 4 labels
print(df.shape)              # see total rows and columns
```

```python
import re, string, spacy

# Load small English model
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])  # faster, only tokenizer + tagger
punct = str.maketrans("", "", string.punctuation)

def clean_text(text):
    text = str(text).lower().translate(punct)             # lowercase + remove punctuation
    text = re.sub(r"\d+", " ", text)                      # remove numbers
    text = re.sub(r"\s+", " ", text).strip()              # collapse whitespace
    doc = nlp(text)
    tokens = [t.lemma_ for t in doc if not (t.is_stop or t.is_punct or t.is_space)]
    return " ".join(tokens)

# ⚠️ Run only on a sample first to test speed
df['clean_text'] = df['text'].head(1000).apply(clean_text)
df[['text', 'clean_text']].head()
```

```python
df.head()
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df['clean_text'], df['label'],
    test_size=0.2, random_state=42, stratify=df['label']
)

print("Train size:", len(X_train))
print("Test size:", len(X_test))
```

```python
# -------------------------------
# 📘 Logistic Regression Classifier
# -------------------------------

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt
import joblib
import os

# ✅ Ensure reports and models folders exist (in parent directory)
os.makedirs("../reports", exist_ok=True)
os.makedirs("../models", exist_ok=True)

# -------------------------------
# TF-IDF Vectorization
# -------------------------------
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)

# -------------------------------
# Train Logistic Regression Model
# -------------------------------
log_model = LogisticRegression(max_iter=2000)
log_model.fit(X_train_tfidf, y_train)
y_pred = log_model.predict(X_test_tfidf)

# -------------------------------
# Evaluate Model
# -------------------------------
acc_log = accuracy_score(y_test, y_pred)
print(f"📘 Logistic Regression Accuracy: {acc_log:.4f}\n")
print(classification_report(y_test, y_pred))

# -------------------------------
# Confusion Matrix
# -------------------------------
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred),
                              display_labels=log_model.classes_)
disp.plot(xticks_rotation=45, cmap='Blues')
plt.title("Logistic Regression - Confusion Matrix")

# ✅ Save confusion matrix one level up in reports folder
plt.savefig("../reports/confusion_matrix_logistic.png", bbox_inches='tight')
plt.show()

# -------------------------------
# Save Model
# -------------------------------
joblib.dump(log_model, '../models/logistic_model.pkl')
print("✅ Logistic Regression model saved to /models folder\n")
```

```python
# -------------------------------
# 📗 Naive Bayes Classifier
# -------------------------------
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt
import joblib
import os

# ✅ Ensure reports and models folders exist (in parent directory)
os.makedirs("../reports", exist_ok=True)
os.makedirs("../models", exist_ok=True)

# -------------------------------
# Train Naive Bayes Model
# -------------------------------
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)

# -------------------------------
# Predictions & Evaluation
# -------------------------------
y_pred_nb = nb_model.predict(X_test_tfidf)

acc_nb = accuracy_score(y_test, y_pred_nb)
print(f"📗 Naive Bayes Accuracy: {acc_nb:.4f}\n")
print(classification_report(y_test, y_pred_nb))

# -------------------------------
# Confusion Matrix
# -------------------------------
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_nb),
                              display_labels=nb_model.classes_)
disp.plot(xticks_rotation=45, cmap='Greens')
plt.title("Naive Bayes - Confusion Matrix")

# ✅ Save confusion matrix in parent /reports folder
plt.savefig("../reports/confusion_matrix_nb.png", bbox_inches='tight')
plt.show()

# -------------------------------
# Save Model
# -------------------------------
joblib.dump(nb_model, '../models/nb_model.pkl')
print("✅ Naive Bayes model saved to /models folder\n")
```

```python
# -------------------------------
# 📙 SVM (LinearSVC) Classifier
# -------------------------------
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
import matplotlib.pyplot as plt
import joblib
import os

# ✅ Ensure reports and models folders exist (in parent directory)
os.makedirs("../reports", exist_ok=True)
os.makedirs("../models", exist_ok=True)

# -------------------------------
# Train SVM Model
# -------------------------------
svm_model = LinearSVC()
svm_model.fit(X_train_tfidf, y_train)

# -------------------------------
# Predictions & Evaluation
# -------------------------------
y_pred_svm = svm_model.predict(X_test_tfidf)

acc_svm = accuracy_score(y_test, y_pred_svm)
print(f"📙 SVM (LinearSVC) Accuracy: {acc_svm:.4f}\n")
print(classification_report(y_test, y_pred_svm))

# -------------------------------
# Confusion Matrix
# -------------------------------
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_svm),
                              display_labels=svm_model.classes_)
disp.plot(xticks_rotation=45, cmap='Oranges')
plt.title("SVM (LinearSVC) - Confusion Matrix")

# ✅ Save confusion matrix in parent /reports folder
plt.savefig("../reports/confusion_matrix_svm.png", bbox_inches='tight')
plt.show()

# -------------------------------
# Save Model
# -------------------------------
joblib.dump(svm_model, '../models/svm_model.pkl')
print("✅ SVM model saved to /models folder\n")
```
